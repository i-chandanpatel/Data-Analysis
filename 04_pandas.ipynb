# Comparing DataFrames

```python
data1 = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Department': ['HR', 'IT', 'Finance']
}
df1 = pd.DataFrame(data1)

data2 = {
    'Name': ['Alice', 'Bob', 'David'],
    'Age': [26, 30, 36],
    'Department': ['HR', 'IT', 'Finance']
}
df2 = pd.DataFrame(data2)
print("\n\nOriginal DataFrame\n\n")
print(df1)
print(df2)
# Set the index to 'Name' for both DataFrames
df1.set_index('Name', inplace=True)
df2.set_index('Name', inplace=True)
print("\n\nDataFrame after setting the index to 'Name' for both DataFrames\n\n")
print(df1)
print(df2)
# Reindex the DataFrames to ensure they have the same index and columns
df1 = df1.reindex(df2.index)
df2 = df2.reindex(df1.index)

# Compare the DataFrames
comparison = df1.compare(df2)
print("\n\n DataFrame after Reindexing\n\n")
print(df1)
print(df2)
print("\n\n After comaprison result")
print(comparison)
```

# Pivoting
#### Pivoting in Python typically refers to reorganizing data in a table-like structure such that rows become columns or vice versa.

##### pivot(): Creates a new DataFrame by rearranging data.
##### pivot_table(): Similar to pivot(), but it allows for aggregation in case of duplicate entries.

```python
dict={"keys":[1,2,3,4],
    "names":["Denver","Lockup","Length","Cever"],
    "houses":["Delhi","London","Shanghai","Ghana"],
    "grades":['A','B','A','B']}

df=pd.DataFrame(dict)
pivot_df=df.pivot(index="keys",columns="names",values=["houses","grades"])
print(pivot_df)
```

```python
dict={"Date":[1,1,3,3],
    "Names":["Denver","Ravi","Ravi","Ravi"],
    "Temp":[10,25,23,12]}
df=pd.DataFrame(dict)
pivot_table=df.pivot_table(index='Date',columns='Names',values='Temp',aggfunc='mean')
print(pivot_table)
```

# Melting
##### Melting typically refers to reshaping data, particularly in the context of working with tabular data using pandas. It is a way to transform data from a wide format to a long format.

```python
# Sample DataFrame in wide format
data = {
    "Name": ["Alice", "Bob"],
    "Math": [85, 90],
    "Science": [95, 80],
    "English": [88, 89],
}

df = pd.DataFrame(data)

# Melting the DataFrame
melted_df = pd.melt(df, id_vars=["Name"], var_name="Subject", value_name="Score")

print("Original DataFrame:\n", df)
print("\nMelted DataFrame:\n", melted_df)
```
