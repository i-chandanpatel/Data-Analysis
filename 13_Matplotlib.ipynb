```python
# Import essential libraries
import pandas as pd          # For data handling (not heavily used here but useful)
import numpy as np           # For numerical operations
import matplotlib.pyplot as plt  # Core plotting library
```

**Matplotlib Basics**

```python
# Generate 11 evenly spaced values between 0 and 5
x = np.linspace(0, 5, 11)

# Create y values (cube of x)
y = x ** 3

# Print values
print("x:", x)
print("y:", y)
```

**Basic Plot**

```python
# Create a simple line plot
plt.plot(x, y)

# Add title and labels
plt.title("Cube Function")
plt.xlabel("X values")
plt.ylabel("Y values (x^3)")

# Display the plot
plt.show()
```

**Subplots in Matplotlib**

```python
# Create multiple plots in a grid (2 rows × 2 columns)

plt.subplot(2, 2, 1)  # position 1
plt.plot(x, y)
plt.title("x vs x^3")

plt.subplot(2, 2, 3)  # position 3
plt.plot(y, x)
plt.title("x^3 vs x")

plt.subplot(2, 2, 2)  # position 2
plt.plot(x**2, x**2)
plt.title("x^2 vs x^2")

plt.subplot(2, 2, 4)  # position 4
plt.plot(x, y)
plt.title("Repeat Plot")

plt.tight_layout()  # Prevent overlap (IMPORTANT addition)
plt.show()
```

**Object Oriented way**

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches
dpi is the dots-per-inch (pixel per inch).

```python
# Create figure with custom size and resolution
fig = plt.figure(figsize=(10, 5), dpi=100)  #figsize=(width, height)

# Add axes (left, bottom, width, height)
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])     #[left,bottom,width,height]

# Plot data
ax.plot(x, y, label="x^3")

# Add labels and title
ax.set_title("Object-Oriented Plot")
ax.set_xlabel("X values")
ax.set_ylabel("Y values")

# Add legend
ax.legend()

# Add grid
ax.grid(True)

# Show plot
plt.show()
```
